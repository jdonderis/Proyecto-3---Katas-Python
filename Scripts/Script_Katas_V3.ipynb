1. Escribe una función que reciba una cadena de texto como parámetro y devuelva un diccionario con las
frecuencias de cada letra en la cadena. Los espacios no deben ser considerados.

```python
def String_to_dictionay(string):
    """
    Función que cuenta las veces que se repite una letra en un string
    Args:
        string (str): Cadena de texto que introduce el usuario 
    Returns:
        Dictionary: Diccionario con la combinación de letra mas numero de veces que se repite.
    """
    #Creamos un diccionario vacío en el que volcaremos la respuesta. 
    diccionario = {} 
    #Eliminamos los posibles espacios dentro del string
    string = string.replace(" ","")
    #Creamos un bucle que suma uno cada vez que encuentra una letra
    for letra in string: 
        if letra in diccionario:
            diccionario [letra] += 1
        else:
            diccionario [letra] = 1 
    return(diccionario) 

String_to_dictionay("HOLA MUNDO")
```

6. Escribe una función que calcule el factorial de un número de manera recursiva.

```python
def factorial (numero):
    """
    Función que calcula el número factorial de un número dado de forma recursiva

    Args: 
    numero (int): Número entero del que se calcula el factorial

    Return:
    Int: Número factorial del número indicado
    """
    #Ponemos una comprobación para que el número no sea negativo
    if numero < 0:
        return ("Solo se pueden calcular factoriales de valores positivos") 
    #Ponemos una comprobación para que el número no sea decimal  
    elif isinstance (numero, float):
        return ("Solo se pueden calcular factoriales de números enteros")
    #Definimos el caso base
    elif numero == 1 or numero == 0:
        return 1   
    #Definimos el caso recursivo
    else:
        return numero * factorial (numero - 1) 
    
factorial (6) 
```

```python
def caracteres_a_letras (caracteres):
    """
    Función que convierte los caracteres en un string en una tupla en la que se muestran los caracteres 
    sin duplicados como mayuscula y minuscula
    Args:
        caracteres (string): Caracteres a convertir

    Returns:
        List: Lista de tuplas con los caracteres que hay en mayúscula y minúscula.
    """    
    #Convertimos todas las letras a minusculas para que Python pueda diferenciarlas y eliminamos duplicados
    letras_minúsculas = set(map(str.lower, caracteres)) 
    #Convertimos el resultado en una tupla convirtiendo cada letra en mayus o minus
    letras_tupla = list(map(lambda letra: (letra.lower(), letra.upper()), letras_minúsculas)) 
    return (letras_tupla)

caracteres_a_letras ("HOLAhola")
```

```python
from functools import reduce #Importamos la función reduce al no estar integrada de inicio en Python
def list_to_number (Numeros):
    """
    Función que obtiene una lista de dígitos y los concatena en un solo número.

    Args:
        Numeros (list): Lista de números a convertir en un solo número.

    Returns:
        Float: Numero concatenado
    """      
    #Creamos un bucle que recorra todos los digitos en la lista
    for numero in Numeros:
        #Caso 1: Números de un solo dígito
        if numero <= 9: 
            #Multiplicamos cada número/operación por 10 y sumamos el siguiente asegurando que no se superponen.
            Dígito = reduce(lambda x, y: x*10+y, Numeros) 
            #Volcamos el resultado 
            return (Dígito) 
        #Caso 2: Números de más de un dígito
        else: 
            #Convertimos la lista a string
            String_numeros = list(map(str, Numeros)) 
            #Concatenamos todos los elementos
            concatenado = reduce(lambda x,y: x+y, String_numeros) 
            #Transformamos a número de nuevo
            Definitivo = int(concatenado) 
            #Volcamos el resultado
            return (Definitivo) 

Lista=[1007, 250, 3]
list_to_number(Lista)         
```

```python
# Paso 1 --> Definimos la función contar_palabras
def contar_palabras(texto):
    """
    Función para contar la cantidad de veces que aparece cada palabra en un texto

    Args:
        texto (str): Texto a analizar.
    Returns:
        Dict: Diccionario que alberga el conteo de cada palabra.
    """ 
    #Creamos un diccionario vacío que albergará el conteo de palabras. 
    conteo = {} 
    #Normalizamos el texto y lo dividimos en palabras.
    texto = texto.lower() 
    #Separamos el texto en palabras.
    palabras = texto.split()
    #Iniciamos un bucle para que se analicen todas las palabras en el texto y sume una al diccionario o al conteo.
    for palabra in palabras: 
        if palabra in conteo:
            conteo[palabra] += 1 
        else:
            conteo[palabra] = 1 
    return conteo
# Paso 2 --> Definimos la función para reemplazar palabras en el texto.
def reemplazar_palabra(texto, palabra, nueva_palabra):
    """
    Función que reemplaza una palabra en el texto por otra indicada.

    Args:
        texto (str): Texto sobre el que realizar el cambio.
        palabra (str): Palabra que será sustituida 
        nueva_palabra (str): Palabra que sustituirá a la palabra indicada.
    Returns: 
        str: Texto con las palabras reemplazadas 
    """   
    #Normalizamos el texto y las palabras para evitar errores de formato de texto 
    texto = texto.lower() 
    palabra = palabra.lower().strip() 
    nueva_palabra= nueva_palabra.lower().strip() 
    #Sustituimos la palabra por la nueva palabra
    texto_nuevo = texto.replace(palabra,nueva_palabra) 
    #Devolvemos el texto con el cambio realizado
    return texto_nuevo 
#Paso 3 --> Definimos la función para eliminar una palabra del texto 
def eliminar_palabra(texto, palabra_borrar):
    """
    Función que elimina una palabra que el usuario elije de un texto

    Args:
        texto (str): Texto a analizar del que eliminaremos la palabra
        palabra_borrar (str): Palabra que eliminaremos
    Returns:
        str: Texto con la palabra eliminada 
    """ 
    #Normalizamos el texto y lo dividimos en palabras
    texto = texto.lower()
    palabras = texto.split()
    #Normalizamos la palabra
    palabra_borrar = palabra_borrar.lower().strip()
    #Iniciamos un bucle que recorra todas las palabras y elimine las palabras
    filtro = []
    for palabra in palabras:
        if palabra_borrar != palabra:
            filtro.append(palabra)
    nuevo_texto = ' '.join(filtro)
    return nuevo_texto

#Paso 4 --> Creamos la función procesar_texto
def procesar_texto(texto, tarea):
    """
    Función utilizada para procesar textos dependiendo de la función que solicite un usuario

    Args:
        texto (str): Texto a procesar 
        tarea (str): Indica la tarea a realizar que puede ser contar_palabras, reemplazar_palabra o eliminar_palabra
    """  
    #Comprobamos que el texto sea un string
    if not isinstance(texto,str): 
        print ("Por favor introduzca un texto no otro tipo de dato")  
    else:
        print ("Bienvenido al procesador de textos, por favor indica que tarea debo realizar")
    #Normalizamos la tarea.
    tarea = tarea.lower().strip()
    #Creamos un primer if para asegurarnos que el usuario ha escrito la función correctamente
   
    #Si indica contar_palabras, lanzamos dicha función
    if tarea == "contar_palabras":
        print(contar_palabras(texto))
    #Si indica reemplazar, cogemos este camino
    elif tarea == "reemplazar_palabra":
        #Solicitamos las palabras
        palabra = input(f"¿Que palabra quieres reemplazar?: ")
        palabra_nueva = input(f"¿Por que palabra la quieres reemplazar?: ")
        #Comprobamos que sean texto 
        if not isinstance(palabra,str):
            print ("Por favor la palabra a sustituir debe ser texto")
        elif not isinstance(palabra_nueva, str):
            print ("Por favor la palabra que utilizaremos debe ser texto")
        else:
            palabra = palabra.lower().strip()
            palabra_nueva = palabra_nueva.lower().strip()
            print(reemplazar_palabra (texto, palabra, palabra_nueva))
    elif tarea == "eliminar_palabra":
         #Solicitamos la palabra
        palabra = input(f"¿Que palabra quieres eliminar?: ")
        #Comprobamos que sean texto 
        if not isinstance(palabra,str): 
            print ("Por favor la palabra a sustituir debe ser texto")
        else:
            print(eliminar_palabra (texto, palabra))
    else:
        if tarea not in ["contar_palabras", "reemplazar_palabra", "eliminar_palabra"]:
            return ("Por favor indique una tarea permitida. Seleccione entre las opciones ofrecidas")
        
procesar_texto ("HOLA MUNDO HOLA MUNDO HOLA MUNDO", "eliminar_palabra")
```

40. Escribe una función que tome dos parámetros: figura (una cadena que puede ser "rectangulo" , "circulo" o  "triangulo" ) y datos (una tupla con los datos necesarios para calcular el área de la figura). 

```python
#Importamos la función pi 
from math import pi
def calcular_areas (figura, datos): 
    """
    Función que calcula el área de una figura (rectángulo, circulo o triangulo) en base a los datos necesarios.
    Args:
        figura (str): Figura para la que calculamos el área
        datos (tuple): Tupla que incluye los datos necesarios para calcular el área. La base y altura en el
        caso del rectángulo y el triángulo y el radio y un 0 en el caso del circulo. (base, altura)
    Returns: 
        float: Indica el area de la figura. 
    """ 
    base, altura = datos
    #Normalizamos la figura para evitar errores case sensitive
    figura_formateada = figura.lower()   
    #Elevamos un error si los números introducidos no son números
    try:
        base = float(base)
        altura = float(altura)
    except ValueError:
        print("Introduzca dos números")
    #Opción 1--> Cálculo del area si la figura es un rectangulo.
    if figura_formateada == "rectangulo":
        area = float(base) * float(altura)
        return (f"El area de la figura {area}")
    #Opción 2--> Cálculo del area si la figura es un triángulo.
    elif figura_formateada == "triangulo":
        area = (float(base) * float(altura)) / 2
        return (f"El area de la figura {area}")
    #Opción 3--> Cálculo del area si la figura es un circulo.
    elif figura_formateada == "circulo":
        area = pi * (float(base)*float(base))
        return (f"El area de la figura {area}")
    #Opción 4 --> Elevamos un error en caso de que no se indique una figura
    elif figura_formateada not in ["rectangulo", "circulo", "triangulo"]:
        return ("Introduzca una figura válida. Rectángulo, circulo o triángulo")

data = (2,2)
calcular_areas("rectangulo", data)
```
